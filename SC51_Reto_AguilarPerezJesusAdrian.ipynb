Construye una gráfica residual de las 3 variables, automóvil (variable dependiente) junto con autobús y camioneta. Las dos últimas son las variables independientes.

```python
X1 = Reto3[['Autobus', 'Camioneta']]
Y1= Reto3['Automovil']
model = LinearRegression()
model.fit(X1, Y1)
y_pred = model.predict(X1)
residuos = Y1 - y_pred
df_residuos = pd.DataFrame({'Automovil': Reto3['Automovil'], 'residuos': residuos})
sns.residplot(x='Automovil', y='residuos', data=df_residuos, color='g')
plt.title('Gráfica Residual para Automóvil')
plt.xlabel('Automóvil')
plt.ylabel('Residuos')
plt.show()
```
